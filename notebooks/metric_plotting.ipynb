# Plot Metrics

```python
def plot_metric_over_epochs(
        metric: str,
        metric_name: str,
        file_names: list[str] = MODEL_LIST, 
        labels: list[str] = LABEL_LIST, 
        colors: list[str] = muted_palette, 
        path_to_results: str = PATH_TO_RESULTS,
        fig_size: tuple = (5, 3.5),
        save_plot: str | None = None
    ) -> None:
    """
    Plot the training loss over epochs for different models.

    Args:

        metric (str): The name of the metric to plot.
        metric_name (str): The name of the metric to display on the y-axis.
        file_names (list[str]): List of filenames containing the results.
        labels (list[str]): List of labels for the models.
        colors (list[str]): List of colors for the plots.
        path_to_results (str): Path to the directory containing the results.
        fig_size (tuple): Size of the figure.
        save_plot (str | None): If provided, the plot will be saved to this path. (e.g "plots/plot.pdf")
    """

    plt.figure(figsize=fig_size)
    plt.xlabel("Epoch")
    plt.ylabel(metric_name)

    columns = ["epoch", "seed", metric]

    for idx, filename in enumerate(file_names):
        tmp = pd.read_csv(path_to_results + filename)

        tmp = tmp[columns].pivot(columns="seed", index="epoch")
        tmp_mean = tmp.mean(axis = 1)
        tmp_std = tmp.std(axis = 1)

        plt.plot(tmp_mean.index +1, tmp_mean, color=colors[idx], label =labels[idx])
        #plt.fill_between(tmp_mean.index+1, tmp_mean - tmp_std, tmp_mean + tmp_std, color=colors[idx], alpha=0.2)
        plt.errorbar(tmp_mean.index +1, tmp_mean, yerr=tmp_std, fmt='o', color=colors[idx], markersize=1, capsize=2, elinewidth=0.5)

    plt.legend()
    if save_plot:
        plt.savefig(save_plot, bbox_inches='tight')
    plt.show()
```

# Plot the ROC curve for all models

```python
def str_to_np(s: pd.Series, row: int = 8) -> np.ndarray:
    """
    Convert a pandas Series of string-encoded arrays into a 2D numpy matrix.
    
    Args:
        s: pandas Series where each element is a string representing an array
        row: index of the row to convert (default is 8)
        
    Returns:
        2D numpy array where each row corresponds to one of the original arrays
    
    """
    cleaned = s.iloc[row].replace("[", " ").replace("]", " ").replace("\n", "")
    array = np.array([float(x) for x in cleaned.split() if x])
    
    return np.vstack(array)
```

```python
def plot_ROC_curve(
        file_names: list[str] = MODEL_LIST, 
        labels: list[str] = LABEL_LIST, 
        colors: list[str] = muted_palette, 
        path_to_results: str = PATH_TO_RESULTS,
        row: int = 4,
        save_plot: str | None = None
        
    ) -> None:
    """
    Plot the ROC curve for different models.

    Args:
        file_names (list[str]): List of filenames containing the results.
        labels (list[str]): List of labels for the models.
        colors (list[str]): List of colors for the plots.
        path_to_results (str): Path to the directory containing the results.
        row (int): Row index to select from the DataFrame.
        save_plot (str | None): If provided, the plot will be saved to this path. (e.g "plots/plot.pdf")
    """

    plt.figure(figsize=(5, 5))
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', linewidth = 0.5, label='Random Guessing')

    for idx, filename in enumerate(file_names):
        tmp = pd.read_csv(path_to_results + filename)
        fpr_axis = str_to_np(tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_fpr"], row)
        tpr_axis = str_to_np(tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_tpr"], row)
        roc_auc = tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_roc_auc"].iloc[row]
        plt.plot(fpr_axis, tpr_axis, color=colors[idx], label=f'{labels[idx]} (AUC = {roc_auc:.3f})')

    plt.legend(loc='lower right', framealpha=0.9)
    plt.legend()
    
    if save_plot:
        plt.savefig(save_plot, bbox_inches='tight')

    plt.show()
```

```python
def plot_all_ROC_curve(
        file_names: list[str] = MODEL_LIST, 
        labels: list[str] = LABEL_LIST, 
        colors: list[str] = muted_palette, 
        path_to_results: str = PATH_TO_RESULTS,
        row: int = 4,
        save_plot: str | None = None,
        ax: plt.Axes = None 
    ) -> None:
    """
    Plot the ROC curve for different models for all seeds.

    Args:
        file_names (list[str]): List of filenames containing the results.
        labels (list[str]): List of labels for the models.
        colors (list[str]): List of colors for the plots.
        path_to_results (str): Path to the directory containing the results.
        row (int): Row index to select from the DataFrame.
        save_plot (str | None): If provided, the plot will be saved to this path. (e.g "plots/plot.pdf")
        ax (plt.Axes): Matplotlib Axes object to plot on. If None, creates a new figure.
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=(5, 5))
    
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.set_title("ROC Curve")

    ax.plot([0, 1], [0, 1], linestyle='--', color='gray', linewidth=0.5, label='Random Guessing')

    for idx, filename in enumerate(file_names):
        tmp = pd.read_csv(path_to_results + filename)
        for row in range(len(tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_fpr"])):
            fpr_axis = str_to_np(tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_fpr"], row)
            tpr_axis = str_to_np(tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_tpr"], row)
            roc_auc = tmp.loc[tmp["epoch"] == tmp["epoch"].max(), "test_roc_auc"].iloc[row]
            if row == 0:
                ax.plot(fpr_axis, tpr_axis, color=colors[idx], label=f'{labels[idx]} (AUC = {roc_auc:.3f})')
            else:
                ax.plot(fpr_axis, tpr_axis, color=colors[idx], alpha=0.2)

    ax.legend(loc='lower right', framealpha=0.9, fontsize=7.5)
    
    if save_plot:
        plt.savefig(save_plot, bbox_inches='tight')

    if ax is None:
        plt.show()
```
